# Q1. Finding Best Location for startup

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file=pd.read_csv("startup_funding.csv")
file=file["CityLocation"]
file.dropna(inplace=True)
a=file[file.str.contains("/")]
l=["Bangalore","bangalore","Mumbai","Delhi","New Delhi","Gurgaon","Noida"]
loc=[]
for i in file:
    a=i.split("/")
    if len(a)>1:
        b=a[0].strip()
        c=a[1].strip()
        if b in l:
            loc.append(b)
        elif c in l:
            loc.append(c)
    else:
        d=a[0].strip()
        if d in l:
            loc.append(d)
        
loc=pd.Series(loc)
loc.replace("bangalore","Bangalore",inplace=True)
loc.replace(["Delhi","New Delhi","Gurgaon","Noida"],"NCR",inplace=True)

ans=loc.value_counts()

city=[]
startups=[]
for j in range(3):
    x=ans.index[j]
    y=ans.values[j]
    city.append(x)
    startups.append(y)
    print(x,y)
    
locations=["Bangalore",]
plt.bar(city,startups,width=0.4,color=["r","g","b"]) 
plt.title("Location vs No. of fundings",fontsize=20)
plt.xlabel("Locations",fontsize=30)
plt.ylabel("No.of fundings",fontsize=20)
plt.show()

```

## Conclusion and Justification

### As we can select a location out of 3 locations (Bangalore, Mumbai and NCR).
The above graph shows comparison between 3 locations(Bangalore, Mumbai and NCR) on
the basis of no. of fundings done by different investors in the regions. 
We can conclude from the graph that the startup is more likely to find an investor in NCR.
This can be due the fact that the location is located near the National capital.
Moreover NCR consists of some major IT hubs namey Gurgaon, New Delhi and Noida.
So, I would suggest my friend to choose NCR for his startup as it increases the chances of finding an investor.

# Q2. Find the top 5 investors who have invested maximum number of times

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

file=pd.read_csv("startup_funding.csv")
name=file["InvestorsName"]

name.dropna(inplace=True)


finallist=[]
def indi(names):
    l=names.split(",")
    for e in l:
        e=e.strip()
        finallist.append(e)
        

name.apply(indi)

f=pd.Series(finallist)
ans=f.value_counts()

Investor=[]
No_investment=[]


for i in range(5):
    a=ans.index[i]
    b=ans.values[i]
    Investor.append(a)
    No_investment.append(b)
    print(ans.index[i],ans.values[i])

    
plt.bar(Investor,No_investment)
plt.title("InvestorName vs No.of Investments",fontsize=15)
plt.xticks(rotation=30)
plt.xlabel("Investor Name",fontsize=20)
plt.ylabel("No. of investments",fontsize=15)
plt.show()
```

## Conclusion and Justification

After failing to find investors we realise we need to work harder and approach the investors directly.
So now we try to analyse the data to find top most investors who have invested the maximum number of times in startup.

The above graph shows a plot between the top 5 most frequent investors and the number of times the investors
invested (including investment in the same startup on multiple stages as unique investments and counting them exclusively).
I have created a list of top 5 investors with maximum number ofinvestments. My friend can use this list to approach these
investors as there are higher chances these investors will invest inhis startup rather than approaching random investors.

# Q3.Find the top 5 investors who have invested maximum number of times in different companies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

file=pd.read_csv("startup_funding.csv")
a=file["InvestorsName"]
b=file["StartupName"]


b.replace("OyoRooms","Oyo",inplace=True)
b.replace("Oyorooms","Oyo",inplace=True)
b.replace("Oyo Rooms","Oyo",inplace=True)
b.replace("OYO Rooms","Oyo",inplace=True)
b.replace("Ola Cabs","Ola",inplace=True)
b.replace("Olacabs","Ola",inplace=True)
b.replace("Flipkart.com","Flipkart",inplace=True)
b.replace("Paytm Marketplace","Paytm",inplace=True)

a.dropna(inplace=True)
b=b[a.index]

b.dropna(inplace=True)
a=a[b.index]

df=pd.DataFrame(a)
df["Startup"]=b

Final=set()

for i in list(df.index):
    ind=i
    start=b[i]
    l=a[i].split(",")
    for e in l:
        individual=e.strip()
        Final.add((individual,start))
        
        
l=list(Final)


inv=[]
for e in l:
    inv.append(e[0])
ans=pd.Series(inv)
ans=ans[ans!=""]

result=ans.value_counts()

for z in range(5):
    print(result.index[z],result.values[z])

plt.bar(result.index[:5],result.values[:5])
plt.xticks(rotation=30)
plt.title(" The top 5 investors who have invested maximum number of times in different companies",fontsize=14)
plt.xlabel("Investor Name",fontsize=15)
plt.ylabel("No. of investments", fontsize =15)
plt.show()


    
    
```

## Conclusion and Justification

It seems all the work we did just go in vain as we were unable to find any investor.
We realised that certain investors may me interested in some particular startups and they
invested in those startups more than one time. Thus the numbers we reached may not give a
good idea about the investing scenario.
Thus, this time we focus on investors which invest in different startups ratherthan investing in same startups
over and over again.

The above graph shows the top 5 investors and number of investments made by them
(reinvesting in same startup is not counted). Hope this time the startup gets investors by contacting these investors.

# Q4. Find the Top 5 investors who have invested in a different number of startups and their investment type is Crowdfunding or Seed Funding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

file=pd.read_csv("startup_funding.csv")
a=file["InvestorsName"]
b=file["StartupName"]
c=file["InvestmentType"]

c.replace("PrivateEquity","Private Equity",inplace=True)
c.replace("SeedFunding","Seed Funding",inplace=True)
c.replace("Crowd funding","Crowd Funding",inplace=True)

b.replace("OyoRooms","Oyo",inplace=True)
b.replace("Oyorooms","Oyo",inplace=True)
b.replace("Oyo Rooms","Oyo",inplace=True)
b.replace("OYO Rooms","Oyo",inplace=True)
b.replace("Ola Cabs","Ola",inplace=True)
b.replace("Olacabs","Ola",inplace=True)
b.replace("Flipkart.com","Flipkart",inplace=True)
b.replace("Paytm Marketplace","Paytm",inplace=True)

a.dropna(inplace=True)
b=b[a.index]
c=c[a.index]

b.dropna(inplace=True)
a=a[b.index]
c=c[b.index]

c.dropna(inplace=True)
a=a[c.index]
b=b[c.index]

df=pd.DataFrame(a)
df["Startup"]=b
df["Type"]=c

Final=set()

for i in list(df.index):
    ind=i
    start=b[i]
    if c[i]=="Crowd Funding" or c[i]=="Seed Funding":
        l=a[i].split(",")
        for e in l:
            individual=e.strip()
            Final.add((individual,start))
        
        
l=list(Final)


inv=[]
for e in l:
    inv.append(e[0])
ans=pd.Series(inv)
ans=ans[ans!=""]
ans=ans[ans!="Undisclosed investors"]
ans=ans[ans!="Undisclosed Investors"]

result=ans.value_counts()

for z in range(6):
    print(result.index[z],result.values[z])

plt.bar(result.index[:6],result.values[:6])
plt.xticks(rotation=90)
plt.title("Top 5 investors whose investment type is Crowdfunding or Seed Funding",fontsize=14) 
plt.xlabel("Investor Name",fontsize=15)
plt.ylabel("No. of investments", fontsize =15)
plt.show()



 
```

## Conclusion and Justification 

It seems we missed some really important knowledge on investment world and so we failed once again.

startups in their initial phase get these kind of fundings only.
So we now compile a list of top 5 investors whose investment type is either Seed Funding or Crowd Funding.
The above bar graph shows top 5 investors of this kind (on X-axis) and the number of times they invested (on Y-axis).
